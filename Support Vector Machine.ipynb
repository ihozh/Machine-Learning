# SVM

```python
import math
import numpy as np
import matplotlib.pyplot as plt


# Create and plot multivariate normal distribution
mean = [0, 0]
cov = [[1,0],[0,1]]
x1, y1 = np.random.multivariate_normal(mean, cov, 100).T/5
x, y = np.random.multivariate_normal(mean, cov, 100).T
label0 = [0 for _ in range(len(x))]

plt.figure(1)
plt.plot(x1, y1, 'x')
plt.axis('equal')

# # Generate g    
# def (xy):
#     res_z = []
#     for z in xy:
#         z = np.array(z)
#         res_z.append(z /10  + z / np.linalg.norm(z))
#     return res_z
# xy = zip(x, y)
# res_z = g(xy)

# Generate z    
def g(xy):
    res_z = []
    for z in xy:
        z = np.array(z)
        res_z.append(z /10  + z / np.linalg.norm(z))
    return res_z
xy = zip(x, y)
res_z = g(xy)

# Plot z
zx, zy = zip(*res_z)
label1 = [1 for _ in range(len(zx))]
# plt.figure(2)
plt.plot(zx, zy, 'o')
```

```python
x = np.concatenate((x1, zx), axis=0)
y = np.concatenate((y1, zy), axis=0)

X_train = np.stack((x, y), axis=1)
y_train = np.concatenate((label0, label1))

from sklearn import svm
# clf = svm.SVR(kernel='rbf', C = 10.0, gamma=0.1)
clf = svm.SVR()
clf.fit(X_train, y_train)
```

```python
# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = X_train[:, 0].min() - .1, X_train[:, 0].max() + .1
y_min, y_max = X_train[:, 1].min() - .1, X_train[:, 1].max() + .1
h = .02  # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
test = np.c_[xx.ravel(), yy.ravel()]
Z = clf.predict(test)
Z = Z.reshape(xx.shape)
```

```python
fig = plt.figure()
ax = plt.axes(xlim=(x_min, x_max), ylim=(y_min, y_max))
plt.pcolormesh(xx, yy, Z, shading='auto', cmap = "bwr", alpha = 0.2)
plt.colorbar()
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, edgecolors='k', cmap=plt.cm.Paired)
# fig.savefig("SVM.pdf", bbox_inches='tight')
```
